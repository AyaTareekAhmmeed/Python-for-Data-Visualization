# What is PyGWalker?


PyGWalker (Python + Graphic Walker) is an interactive data exploration tool designed to work directly inside Jupyter or Colab notebooks. It's like Tableau or Power BI, but built for Python users — no drag-and-drop UI needed outside your notebook.

It transforms your Pandas DataFrame into a rich, visual exploration UI — allowing you to:

Slice and dice data with charts.

Create pivot tables.

Explore patterns with filters and aggregations.

Export charts or insights easily.

It’s built on top of Graphic Walker (an open-source visualization tool powered by Vega-Lite).



# Why use PyGWalker?


✅ Tableau-style exploration without leaving your notebook

✅ Works with pandas.DataFrame instantly — no ETL or upload

✅ Great for data profiling, EDA, and insight discovery

✅ Compatible with Jupyter, VS Code, and Streamlit

✅ Free and open-source
